```python
data.rename(columns={'Diabetes_012':'Diabetes'},inplace=True)
data.head()
```

```python
# Created a function to get necessary information from the data set in a dataframe
def get_data_info():
    dataa = pd.DataFrame(index=data.columns)
    dataa['Data_type'] = data.dtypes
    dataa['Total Value'] = data.count()
    dataa['Null_count'] = data.isnull().sum()
    dataa['Unique_count'] = data.nunique()
    return dataa
```

```python
get_data_info()
```

```python
#It is now easy to view the missing data in all fields, their unique values, datatypes and Null Value Count
#The data doesn't have any Null Values. So let's check for outliers and remove them using z-score.
```

```python
#Z-Score indicates how many standard deviations away a data point is. Then you remove those points that are 3 standard 
#deviations away. 
```

```python
columnstoremoveoutliers = ['BMI','Age','MentHlth','PhysHlth']
data[columnstoremoveoutliers].describe()
```

```python

for i in columnstoremoveoutliers:
    z = 'z_score' + i
    data[z] = (data[i] - data[i].mean())/data[i].std()
```

```python
data.head()
```

```python
for i in columnstoremoveoutliers:
    z = 'z_score'+i
    data = data[(data[z]>-3) | (data[z]<3)]
get_data_info()
```

```python
data.Diabetes.value_counts()
```

```python
#All outliers have been cleaned. 
```

# DATA VISUALIZATION

```python
palette3 = ['#AAB607','#FA6147','#FA6147']

plt.figure(figsize=(10,7.5))
label = ['Do not have Diabetes','Prediabetes','Have Diabetes']
patches, texts, pcts = plt.pie(data.Diabetes.value_counts(),
                               labels=label,
                               colors=['#AAB607','#4BEEE6','#FA6147'],
                               pctdistance=0.80,
                               shadow=False,
                               startangle=5,
                               autopct='%1.1f%%',
                               textprops={'fontsize': 15,
                                          'weight': 'bold'
                                          })
plt.setp(pcts, color='white')

hfont = {'fontname':'calibri', 'weight': 'bold'}
plt.title('Percentage of People living with Diabetes', size=25, **hfont)

centre_circle = plt.Circle((0,0),0.65,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

```python
DiabetesHighBPDf = data.groupby(['Diabetes','HighBP']).size().reset_index(name ='Count')
# DiabetesHighBPDf = pd.DataFrame(DiabetesHighBPDf).reset_index(name ='Count')
DiabetesHighBPDf
```

```python
plt.figure(figsize = (8,6))

g = sns.barplot(x = 'Diabetes', y = 'Count', hue = 'HighBP', data = DiabetesHighBPDf, palette = 'Set2')
g.legend(labels=['Do not have BP','Have BP'])
g.set_xticks(range(len(DiabetesHighBPDf['Diabetes'].unique()))) # <--- set the ticks first
g.set_xticklabels(['Healthy','Diabetes','Prediabetic'])
plt.title("Dibaetes Status ~ BP")
plt.show()
```
